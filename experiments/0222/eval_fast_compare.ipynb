```python
import gc, json, time, math, random, traceback, os, sys, platform
from pathlib import Path
import pandas as pd
import torch
from transformers import AutoTokenizer, AutoModelForCausalLM

print("python:", sys.executable)
print("cwd:", os.getcwd())
print("platform:", platform.platform())
print("cuda available:", torch.cuda.is_available())
```

```python
BASE_0225 = Path(r"C:\Users\FORYOUCOM\Downloads\testdata\0225\models")
BASE_0213 = Path(r"C:\Users\FORYOUCOM\Downloads\testdata\0213\models")
EVAL_JSONL = Path(r"C:\Users\FORYOUCOM\Downloads\testdata\0222\eval_samples_official_512.jsonl")
OUT_DIR = Path(r"C:\Users\FORYOUCOM\Downloads\testdata\0222\eval")
OUT_DIR.mkdir(parents=True, exist_ok=True)

MODEL_DIRS = {
    "A_mix50_50": BASE_0225 / "out_q_0225_mix50_50",
    "B_mix30_70": BASE_0225 / "out_q_0225_mix30_70",
    "C_mix70_30": BASE_0225 / "out_q_0225_mix70_30",
    "D_keep_0_3_26": BASE_0213 / "out_q_keep_0_3_26",
    "E_keep_0_3_26_stratified": BASE_0213 / "out_q_keep_0_3_26_stratified",
    "F_official_keep_0_4_26_30": BASE_0213 / "out_q_official_keep_0_4_26_30",
}

NUM_SAMPLES = 32
SEED = 42
MAX_SEQ_LEN_FOR_PPL = 64

assert EVAL_JSONL.exists(), f"missing eval file: {EVAL_JSONL}"
for k, v in MODEL_DIRS.items():
    assert v.exists(), f"missing model ({k}): {v}"

print("Model paths OK")
```

```python
def load_prompts(path: Path, n=32, seed=42):
    texts = []
    with open(path, "r", encoding="utf-8") as f:
        for line in f:
            if line.strip():
                obj = json.loads(line)
                t = obj.get("text", "")
                if t:
                    texts.append(t)
    random.Random(seed).shuffle(texts)
    return texts[:n]

def compute_ppl_proxy(model_dir: Path, texts, max_seq_len=64):
    tok = AutoTokenizer.from_pretrained(
        str(model_dir),
        local_files_only=True,
        use_fast=False,
        trust_remote_code=True
    )

    model = AutoModelForCausalLM.from_pretrained(
        str(model_dir),
        torch_dtype=torch.float16 if torch.cuda.is_available() else torch.float32,
        device_map="auto" if torch.cuda.is_available() else None,
        local_files_only=True,
    )
    model.eval()

    losses = []
    with torch.no_grad():
        for t in texts:
            enc = tok(t, return_tensors="pt", truncation=True, max_length=max_seq_len)
            if torch.cuda.is_available():
                ids = enc["input_ids"].to(model.device)
                attn = enc["attention_mask"].to(model.device)
            else:
                ids = enc["input_ids"]
                attn = enc["attention_mask"]
            out = model(input_ids=ids, attention_mask=attn, labels=ids)
            losses.append(float(out.loss.detach().cpu()))

    del model
    gc.collect()
    if torch.cuda.is_available():
        torch.cuda.empty_cache()

    mean_nll = sum(losses) / max(len(losses), 1)
    return float(math.exp(mean_nll))
```

```python
PROMPTS = load_prompts(EVAL_JSONL, NUM_SAMPLES, SEED)
print("Loaded prompts:", len(PROMPTS))

rows = []
for model_name, model_dir in MODEL_DIRS.items():
    row = {"model_name": model_name, "model_dir": str(model_dir)}
    try:
        t0 = time.perf_counter()
        ppl = compute_ppl_proxy(model_dir, PROMPTS, MAX_SEQ_LEN_FOR_PPL)
        t1 = time.perf_counter()

        row.update({
            "status": "success",
            "ppl_proxy": round(ppl, 8),
            "elapsed_sec": round(float(t1 - t0), 4),
            "error": "",
        })
        print(f"[OK] {model_name}: ppl={ppl:.6f}, sec={t1-t0:.2f}")
    except Exception as e:
        row.update({
            "status": "failed",
            "ppl_proxy": None,
            "elapsed_sec": None,
            "error": f"{type(e).__name__}: {str(e)}",
        })
        print("[ERROR]", model_name, row["error"])
        print(traceback.format_exc(limit=2))
    rows.append(row)

res_df = pd.DataFrame(rows).sort_values(["status", "ppl_proxy"], ascending=[True, True])
display(res_df)
res_path = OUT_DIR / "fast_local_ppl_compare.csv"
res_df.to_csv(res_path, index=False, encoding="utf-8-sig")
print("saved:", res_path)
```
